# Prepare data

Choose features for training

```
data.columns = ["vendor name", "Model Name", "MYCT", "MMIN", "MMAX", "CACH", 'CHMIN', 'CHMAX', 'PRP', 'ERP']
data.head()
```

```
features = data.drop("ERP",1)
features.head()
```

```
value = data["ERP"]
```

```
value.head()
```

```
features = pd.get_dummies(features)
features
```

Slipt data into train set and test set

```
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(features, value, test_size=0.30, random_state=1)
```

# Decision Tree

```
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import accuracy_score
```

```
# Create decision tree regressor object
model1 = DecisionTreeRegressor()
# Train model
model1.fit(X_train, y_train)
```

```
print("The Train/ Score is: ", model1.score(X_train,y_train)*100,"%")
print("The Test/ Score accuracy is: ", model1.score(X_test,y_test)*100,"%")
```

# Random forest

```
from sklearn.ensemble import RandomForestRegressor
```

```
rrf1 = RandomForestRegressor(n_estimators=155)
# Train model
model = rrf1.fit(X_train, y_train)
```

```
print("The Train/ Score is: ", model.score(X_train,y_train)*100,"%")
print("The Test/ Score accuracy is: ", model.score(X_test,y_test)*100,"%")
```
